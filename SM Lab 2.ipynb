```python
s1 = [6,7,10,12,13,4,8,12]
a=0
for i in s1:
  a=a+i
men = a/len(s1)
print("Mean = ",men)

lis2 = []
v=0
for j in s1:
  v=(j-men)**2
  lis2.append(v)
print(lis2)
sm=0
for x in lis2:
  sm = sm+x
print(sm)
print(len(lis2))
var = sm/len(lis2)
print("Variance = ",var)
```
