```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
da = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SM/Data.csv')
data = pd.DataFrame(da)
data.head()
```

```python
duplicateRowsDF = data[data.duplicated()]
duplicateRowsDF
```

```python
df = data.drop_duplicates()
```

```python
duplicateRowsDF = df[df.duplicated()]
duplicateRowsDF
```

```python
data.isnull().sum()
```

```python
imputer = SimpleImputer(strategy='mean', missing_values=np.nan)
imputer = imputer.fit(data[['minute']])
data['minute'] = imputer.transform(data[['minute']])
```

```python
imputer = SimpleImputer(strategy='most_frequent', missing_values=np.nan)
imputer = imputer.fit(data[['scorer']])
data['scorer'] = imputer.transform(data[['scorer']])
```
