```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
x1 = np.random.normal(0, 1, 100)
y1 = 2*x1 + np.random.normal(0, 1, 100)
x2 = np.random.normal(0, 1, 100)
y2 = 3*x2 + np.random.normal(0, 1, 100)

# Create a scatter plot of the data
plt.scatter(x1, y1, label='Dataset 1')
plt.scatter(x2, y2, label='Dataset 2')

# Calculate the regression line for each dataset
slope1, intercept1 = np.polyfit(x1, y1, 1)
slope2, intercept2 = np.polyfit(x2, y2, 1)

# Plot the regression lines
plt.plot(x1, slope1*x1 + intercept1, color='blue', label='Regression line 1')
plt.plot(x2, slope2*x2 + intercept2, color='orange', label='Regression line 2')

# Add a legend and axis labels
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')

# Display the plot
plt.show()
```
